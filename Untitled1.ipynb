```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# circle 80
circle80_ga = pd.read_csv("output/long_ga_circle80_mut10.txt",header=None)
circle80_hc = pd.read_csv("output/long_hc_circle80.txt",header=None).T[0:-1].T
circle80_rs = pd.read_csv("output/long_rs_circle80.txt",header=None).T[0:-1].T
circle80_ga_2 = pd.read_csv("output/long_ga_circle80_mut5.txt",header=None)


```

```python
ga_mean = circle80_ga.mean()
ga_std = circle80_ga.std()
ga2_mean = circle80_ga_2.mean()
ga2_std = circle80_ga_2.std()
rs_mean = circle80_rs.mean()
rs_std = circle80_rs.std()
hc_mean = np.zeros(5000); hc_std = np.zeros(5000)
mean = circle80_hc.mean(); std = circle80_hc.var()/2
for i in range(5000):
    if i < 1500:
        hc_mean[i] = mean[i]; hc_std[i] = std[i]
    else:
        hc_mean[i] = mean[1499]; hc_std[i] = std[1499]
```

```python
plt.figure(figsize=(6,4))
ax = plt.subplot(1,1,1)
#ax.plot(np.arange(5000), ga)
ax.errorbar(np.arange(5000), hc_mean, yerr=hc_std, errorevery=500,label="Hill Climber")
ax.errorbar(np.arange(5000), rs_mean, yerr=rs_std, errorevery=500,label="Random Search")
ax.errorbar(np.arange(5000), ga_mean, yerr=ga_std, errorevery=500, label="GA, mutate_rate=0.01", color = 'k')
ax.errorbar(np.arange(5000), ga2_mean, yerr=ga2_std, errorevery=500, label="GA, mutate_rate=0.005", color='r')


ax.set_xlim(0,5000)
ax.set_xlabel("Evaluation")
ax.set_ylabel("Fitness")
#ax.set_title("TSP Longest, N=80")
plt.legend(loc='lower right')
plt.savefig("output/learningcurve1.png", dpi=300)
```
